Student Data Analysis


```python
import pandas as pd

data = {
    "Name": ["Yash", "Rahul", "Aman", "Priya"],
    "Marks": [85, 92, 78, 88]
}

df = pd.DataFrame(data)

# Average marks
avg = df["Marks"].mean()

# Topper
topper = df.loc[df["Marks"].idxmax()]

print("Average Marks:", avg)
print("Topper:", topper["Name"])
```
